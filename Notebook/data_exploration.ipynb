```python
print(df['product_type'].value_counts())
print(df['skintype'].value_counts())
df['brand'].value_counts()
```

```python
df['skintype_list'] = df['skintype'].str.split(',')
```

```python
df['skintype_list'] = df['skintype_list'].apply(lambda x: [s.strip() for s in x])
```

```python
df[['skintype', 'skintype_list']].head()
```

```python
df.drop(columns=['skintype'], inplace=True)
```

```python
df.head()
```

```python
df['notable_effects'].head()
```

```python
df['notable_effects_list'] = df['notable_effects'].str.split(',').apply(lambda x: [s.strip() for s in x])
```

```python
df.drop(columns=['notable_effects'], inplace=True)
```

```python
df['notable_effects_list']
```

```python
df.head(5)
```

```python
df.drop(columns=['Unnamed: 0', 'labels'], inplace=True)
```

```python
df['price'] = df['price'].str.replace(r'[^\d.]', '', regex=True)
```

```python
df.head(5)
```

```python
df.head(5)
```

```python
df[df.duplicated(subset=['product_name', 'product_type'], keep=False)]
```

```python
df.drop_duplicates(subset=['product_name', 'product_type'], inplace=True)
```

Visualizations

```python
# Boxplot to check for outliers
plt.figure(figsize=(10, 2))  # wider and shorter boxplot
sns.boxplot(x=df['price'])
plt.title('Price Outliers')
plt.show()
```

```python

# Bar plot for a categorical field 
plt.figure(figsize=(10, 2)) 
counts = df['product_type'].value_counts()
counts.plot(kind='bar')
plt.title('Count of Product Types')
print(counts.index)  
print(counts.values)
plt.show()
```
