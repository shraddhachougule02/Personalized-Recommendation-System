```python
df['skintype_list'] = df['skintype'].str.split(',')
```

```python
df['skintype_list'] = df['skintype_list'].apply(lambda x: [s.strip() for s in x])
```

```python
df['notable_effects_list'] = df['notable_effects'].str.split(',').apply(lambda x: [s.strip() for s in x])
```

```python
df.drop(columns=['notable_effects'], inplace=True)
```

```python
df['notable_effects_list']
```

```python
df.drop(columns=['Unnamed: 0', 'labels'], inplace=True)
```

```python
df['price'] = df['price'].str.replace(r'[^\d]', '', regex=True)
```

```python
df['price'] =df['price'].astype(float).round().astype(int)
```

```python
df.drop_duplicates(subset=['product_name', 'product_type'], inplace=True)
```

Visualizations

```python
# Boxplot to check for outliers
plt.figure(figsize=(10, 2))  # wider and shorter boxplot
sns.boxplot(x=df['price'])
plt.title('Price Outliers')
plt.show()
```

```python

# Bar plot for a categorical field 
plt.figure(figsize=(10, 2)) 
counts = df['product_type'].value_counts()
counts.plot(kind='bar')
plt.title('Count of Product Types')
print(counts.index)  
print(counts.values)
plt.show()
```

```python
df = df.reset_index(drop=True)
df['product_id'] = 'prod_' + df.index.astype(str)
```

```python
df.to_csv('../Data/cleaned_data.csv', index=False)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your product dataframe with a 'product_type' column
plt.figure(figsize=(10,6))
sns.countplot(y='product_type', data=df, order=df['product_type'].value_counts().index, palette='viridis')
plt.title('Product Distribution by Type')
plt.xlabel('Count')
plt.ylabel('Product Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Product Prices')
plt.xlabel('Price (₹)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

```python
# If you have separate columns as flags for skin types (e.g., 'Oily', 'Dry' etc.)
skin_types = ['Sensitive', 'Oily', 'Dry', 'Combination', 'Normal']
counts = [df[skin].sum() for skin in skin_types]

plt.figure(figsize=(8,5))
sns.barplot(x=skin_types, y=counts, palette='muted')
plt.title('Number of Products by Skin Type Compatibility')
plt.xlabel('Skin Type')
plt.ylabel('Number of Compatible Products')
plt.tight_layout()
plt.show()
```
